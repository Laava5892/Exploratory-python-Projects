```python
import gender_guesser.detector as gender

d = gender.Detector()

print(d.get_gender(u"Bob"))
    

print(d.get_gender(u"Sally"))

print(d.get_gender(u"Laavanya"))

print(d.get_gender(u"Daniel"))

import tardis

bp = tardis.get_bp_postgres_connection()

worker_names = bp.execute_query(
"""
SELECT u.id as worker_id,
name_first,
name_last
FROM public.user u
WHERE u.id in (
    SELECT worker_id
    FROM public.shift_assignment_outcome sao
    WHERE is_official
    AND outcome='Worked'
) 
""")
```

```python
worker_names['inferred_gender'] = worker_names.name_first.apply(d.get_gender)
```

```python
worker_names.inferred_gender.value_counts()
```
